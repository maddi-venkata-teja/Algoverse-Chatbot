```python
import os
import json
import torch
from datasets import Dataset
from unsloth import FastLanguageModel
from trl import SFTTrainer
from transformers import TrainingArguments

# =====================================================================
# STEP 3.1: Load and parse DSA data
# =====================================================================
print("[1/5] Parsing DSA dataset...")

with open("dsa.json", "r") as f:
    raw_data = json.load(f)

formatted_data = []
for concept in raw_data["dsa_concepts"]:
    instruction = f"Explain: {concept['name']}"
    response = f"{concept['name']} is {concept['category']}. {concept['definition']}\n"
    response += f"Complexity: {', '.join([f'{k}:{v['time_complexity']}' for k,v in concept['operations'].items()])}\n"
    response += f"Space: {concept['space_complexity']}\n"
    response += f"Uses: {', '.join(concept['applications'][:2])}"

    formatted_data.append({"text": f"Q: {instruction}\nA: {response}"})

dataset = Dataset.from_dict({"text": [item["text"] for item in formatted_data]})
print(f"✓ Loaded {len(formatted_data)} DSA concepts")

# =====================================================================
# STEP 3.2: Load Phi-3-mini (4-bit)
# =====================================================================
print("[2/5] Loading Phi-3-mini (4-bit)...")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Phi-3-mini-4k-instruct-bnb-4bit",
    max_seq_length=4096,
    load_in_4bit=True,
    dtype=torch.float16
)
print("✓ Model loaded")

# =====================================================================
# STEP 3.3: Apply LoRA
# =====================================================================
print("[3/5] Applying LoRA...")

model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.0,
    bias="none",
    use_gradient_checkpointing="unsloth"
)
print("✓ LoRA applied")

# =====================================================================
# STEP 3.4: Train
# =====================================================================
print("[4/5] Running training...")

training_args = TrainingArguments(
    output_dir="/tmp/dsa_model",
    num_train_epochs=1,
    per_device_train_batch_size=4,
    learning_rate=2e-4,
    logging_steps=1,
    optim="adamw_8bit",
    save_strategy="no",  # Don't save checkpoints
)

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    args=training_args,
    max_seq_length=4096,
)

trainer.train()
print("✓ Training complete")

# =====================================================================
# STEP 3.5: Inference
# =====================================================================
print("[5/5] Setting up chatbot...")

FastLanguageModel.for_inference(model)

def chat(user_input):
    prompt = f"Q: Explain: {user_input}\nA:"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=200, temperature=0.7)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\n" + "="*60)
print("DSA Chatbot Ready - Type 'quit' to exit")
print("="*60 + "\n")

while True:
    inp = input("Ask (e.g., Stack, Queue, Array): ").strip()
    if inp.lower() == "quit":
        break
    if inp:
        print(f"Bot: {chat(inp)}\n")
```
