```python
#week 4
import os
import ast
import sys
import json
import time
import signal
import IPython
import statistics
from io import StringIO
from datetime import datetime
from contextlib import contextmanager
from google.colab import output

# Ensure deployment tracking blueprints exist
os.makedirs("./deployment_repo/model_metadata", exist_ok=True)
os.makedirs("./deployment_repo/model_binary", exist_ok=True)

# =====================================================================
# 1. LOAD CANDIDATE CODE SAMPLES (Week 3 inference outputs -> records)
# =====================================================================
print("[1/5] Restructuring inference outputs into a sanity-check record set...")

dataset_filename = "./deployment_repo/model_metadata/inference_test_report.jsonl"
candidate_records = []
if os.path.exists(dataset_filename):
    with open(dataset_filename, "r") as f:
        for line in f:
            row = json.loads(line)
            if row.get("output"):
                candidate_records.append({
                    "id": row["id"], "category": row["category"],
                    "prompt": row["prompt"], "code": row["output"],
                })

# Fallback fixture set so this cell is self-contained if Week 3 hasn't run yet
if not candidate_records:
    candidate_records = [
        {"id": "C01", "category": "easy",
         "prompt": "factorial",
         "code": "def factorial(n):\n    if n <= 1:\n        return 1\n    return n * factorial(n - 1)\n\nprint(factorial(5))"},
        {"id": "C02", "category": "easy",
         "prompt": "broken factorial (missing colon)",
         "code": "def factorial(n)\n    if n <= 1:\n        return 1\n    return n * factorial(n - 1)\n\nprint(factorial(5))"},
        {"id": "C03", "category": "hard",
         "prompt": "missing base case",
         "code": "def no_base(n):\n    return n * no_base(n - 1)\n\nprint(no_base(5))"},
    ]

print(f"  Loaded {len(candidate_records)} candidate code samples for sanity checking.")

# =====================================================================
# 2. SYNTAX + EXECUTION + RECURSION-SAFETY CHECKS
# =====================================================================
print("[2/5] Running syntax validation and guarded execution...")

sys.setrecursionlimit(200)   # low limit on purpose — surfaces runaway recursion fast

@contextmanager
def time_limit(seconds: int):
    def _handler(signum, frame):
        raise TimeoutError(f"Execution exceeded {seconds}s")
    try:
        signal.signal(signal.SIGALRM, _handler)
        signal.alarm(seconds)
        yield
    finally:
        signal.alarm(0)

def check_syntax(code: str) -> dict:
    try:
        ast.parse(code)
        return {"valid": True, "error": None}
    except SyntaxError as e:
        return {"valid": False, "error": f"L{e.lineno}: {e.msg}"}

def execute_code(code: str, timeout_s: int = 3) -> dict:
    syn = check_syntax(code)
    if not syn["valid"]:
        return {"success": False, "output": "", "error": syn["error"], "exec_ms": 0}

    stdout_capture = StringIO()
    namespace = {}
    start = time.time()
    try:
        with time_limit(timeout_s):
            old_stdout = sys.stdout
            sys.stdout = stdout_capture
            exec(code, namespace)
            sys.stdout = old_stdout
        return {"success": True, "output": stdout_capture.getvalue().strip(),
                "error": None, "exec_ms": (time.time() - start) * 1000}
    except RecursionError:
        sys.stdout = sys.__stdout__
        return {"success": False, "output": "", "error": "RecursionError — infinite recursion detected",
                "exec_ms": (time.time() - start) * 1000}
    except TimeoutError as e:
        sys.stdout = sys.__stdout__
        return {"success": False, "output": "", "error": str(e), "exec_ms": timeout_s * 1000}
    except Exception as e:
        sys.stdout = sys.__stdout__
        return {"success": False, "output": "", "error": f"{type(e).__name__}: {e}",
                "exec_ms": (time.time() - start) * 1000}
    finally:
        sys.stdout = sys.__stdout__

sys.setrecursionlimit(500)

sanity_reports = []
for rec in candidate_records:
    syn = check_syntax(rec["code"])
    run = execute_code(rec["code"])
    report = {
        "id": rec["id"], "category": rec["category"], "prompt": rec["prompt"],
        "syntax_valid": syn["valid"], "executes": run["success"],
        "exec_output": run["output"], "error": run["error"] or syn["error"],
        "exec_ms": run["exec_ms"],
    }
    sanity_reports.append(report)
    icon = "✔" if report["executes"] else "✗"
    print(f"  [{rec['id']}] {icon}  syntax={'OK' if syn['valid'] else 'FAIL'}  "
          f"exec={'OK' if run['success'] else 'FAIL'}  "
          f"{'' if run['success'] else '(' + str(report['error']) + ')'}")

# =====================================================================
# 3. PEP8 STYLE + CYCLOMATIC COMPLEXITY (best-effort, library-gated)
# =====================================================================
print("[3/5] Running style and complexity checks (where libraries are available)...")

try:
    import pycodestyle, tempfile
    def style_clean(code: str) -> bool:
        with tempfile.NamedTemporaryFile(mode="w", suffix=".py", delete=False) as f:
            f.write(code); tmp_path = f.name
        checker = pycodestyle.Checker(tmp_path, show_source=False, show_pep8=False)
        checker.check_all()
        os.unlink(tmp_path)
        return len(checker.results) if hasattr(checker, "results") else 0
    style_available = True
except ImportError:
    style_available = False

try:
    from radon.complexity import cc_visit
    radon_available = True
except ImportError:
    radon_available = False

for report in sanity_reports:
    code = next(r["code"] for r in candidate_records if r["id"] == report["id"])
    if style_available and report["syntax_valid"]:
        try:
            report["pep8_violations"] = style_clean(code)
        except Exception:
            report["pep8_violations"] = None
    else:
        report["pep8_violations"] = None

    if radon_available and report["syntax_valid"]:
        try:
            blocks = cc_visit(code)
            report["max_complexity"] = max((b.complexity for b in blocks), default=1)
        except Exception:
            report["max_complexity"] = None
    else:
        report["max_complexity"] = None

print(f"  pycodestyle available: {style_available}  |  radon available: {radon_available}")

# =====================================================================
# 4. EXPORT CONSOLIDATED SANITY REPORT
# =====================================================================
print("[4/5] Exporting consolidated sanity report...")

REPORT_PATH = "./deployment_repo/model_metadata/code_sanity_report.jsonl"
with open(REPORT_PATH, "w", encoding="utf-8") as f:
    for report in sanity_reports:
        record = {
            **report,
            "metadata": {
                "student": "V.NANI", "roll_no": "A24126552185",
                "week": "Week 4 — Delivery", "task": "Code Sanity Checks",
                "created": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            }
        }
        f.write(json.dumps(record, ensure_ascii=False) + "\n")

pass_rate = sum(1 for r in sanity_reports if r["executes"]) / max(len(sanity_reports), 1)
print(f"  Report exported: {REPORT_PATH}")
print(f"  Execution pass rate: {pass_rate:.0%}")

# =====================================================================
# 5. DEFINE KERNEL-LEVEL SANITY-CHECK BRIDGE (NO BACKEND CODE NEEDED)
# =====================================================================
print("[5/5] Binding native Colab execution hooks...")

def colab_sanity_bridge(user_msg):
    """Runs a live sanity check on pasted code, directly inside the current notebook runtime memory space."""
    code = user_msg
    syn  = check_syntax(code)
    run  = execute_code(code)

    lines = [
        f"Syntax valid : {'YES' if syn['valid'] else 'NO — ' + str(syn['error'])}",
        f"Executes     : {'YES' if run['success'] else 'NO'}",
    ]
    if run["success"]:
        lines.append(f"Stdout       : {run['output'] or '(no output)'}")
        lines.append(f"Exec time    : {run['exec_ms']:.1f} ms")
    elif run["error"]:
        lines.append(f"Error        : {run['error']}")

    clean_reply = "\n".join(lines)
    return IPython.display.JSON({"response": clean_reply})

output.register_callback('notebook.execute_llm', colab_sanity_bridge)

# =====================================================================
# 6. INJECT INDEPENDENT LOCAL FRONTEND WORKSPACE
# =====================================================================
print("Displaying user interface...")

chat_ui_html = """
<div id="dsa-app-wrapper">
    <div class="chat-header">
        <h2>Code Sanity Check Console</h2>
        <p>Paste any Python snippet (multi-line) — press <strong>Ctrl+Enter</strong> or click <strong>Check</strong> to run.</p>
    </div>
    <div class="message-area" id="msgArea">
        <div class="msg bot-msg">System Online. Paste Python code to trigger a live syntax + execution check.</div>
    </div>
    <div class="input-area">
        <textarea id="userInput" placeholder="Paste your Python code here (multi-line supported)..." rows="5" onkeydown="if(event.ctrlKey && event.key === 'Enter') sendQuery()"></textarea>
        <button id="sendBtn" onclick="sendQuery()">Check</button>
    </div>
</div>

<style>
    #dsa-app-wrapper {
        background: #1e293b; color: #f8fafc; border-radius: 12px; max-width: 800px;
        margin: 20px auto; padding: 20px; display: flex; flex-direction: column; height: 550px;
        font-family: system-ui, sans-serif; border: 1px solid #334155; box-shadow: 0 4px 15px rgba(0,0,0,0.2);
    }
    .chat-header h2 { margin: 0; font-size: 1.4rem; color: #3b82f6; }
    .chat-header p { margin: 5px 0; font-size: 0.85rem; color: #94a3b8; }
    .message-area { flex: 1; padding: 15px; overflow-y: auto; display: flex; flex-direction: column; gap: 12px; background: #0f172a; border-radius: 8px; }
    .msg { max-width: 85%; padding: 12px; border-radius: 8px; font-size: 0.95rem; line-height: 1.6; white-space: pre-wrap; }
    .user-msg { background: #3b82f6; color: white; align-self: flex-end; border-bottom-right-radius: 0px; }
    .bot-msg { background: #334155; color: #f8fafc; align-self: flex-start; border: 1px solid #475569; border-bottom-left-radius: 0px; }
    .input-area { margin-top: 15px; display: flex; gap: 10px; }
    #userInput { flex: 1; background: #0f172a; border: 1px solid #475569; border-radius: 6px; padding: 12px; color: white; outline: none; font-size: 0.9rem; font-family: monospace; resize: vertical; min-height: 90px; }
    #userInput:focus { border-color: #3b82f6; }
    #sendBtn { background: #3b82f6; color: white; border: none; padding: 12px 24px; border-radius: 6px; cursor: pointer; font-weight: bold; font-size: 0.95rem; align-self: flex-end; }
    #sendBtn:hover { background: #2563eb; }
</style>

<script>
    async function sendQuery() {
        const inputField = document.getElementById("userInput");
        const sendBtn = document.getElementById("sendBtn");
        const msgArea = document.getElementById("msgArea");
        const queryText = inputField.value.trim();
        if (!queryText) return;

        appendBubble(queryText, "user-msg");
        inputField.value = "";
        inputField.disabled = true; sendBtn.disabled = true;

        const loadingId = appendBubble("Parsing and executing in a guarded sandbox...", "bot-msg");

        try {
            google.colab.kernel.invokeFunction('notebook.execute_llm', [queryText], {}).then(function(result) {
                const responseData = result.data['application/json'];
                document.getElementById(loadingId).innerText = responseData.response;
            });
        } catch (error) {
            document.getElementById(loadingId).innerText = "Runtime connection error. Unable to reach sanity-check bridge.";
        } finally {
            inputField.disabled = false; sendBtn.disabled = false;
            inputField.focus();
            msgArea.scrollTop = msgArea.scrollHeight;
        }
    }

    function appendBubble(text, className) {
        const msgArea = document.getElementById("msgArea");
        const div = document.createElement("div");
        const id = "m_" + Date.now() + Math.floor(Math.random()*1000);
        div.id = id; div.className = "msg " + className; div.innerText = text;
        msgArea.appendChild(div);
        msgArea.scrollTop = msgArea.scrollHeight;
        return id;
    }
</script>
"""

IPython.display.display(IPython.display.HTML(chat_ui_html))
print("\nInterface active above. Use the check pane directly without network setup requirements.")
```

```python
#week1 vector optimisation
import os
import json
import time
import IPython
import numpy as np
from google.colab import output
!pip install faiss-cpu -q
# Ensure deployment tracking blueprints exist
os.makedirs("./deployment_repo/model_metadata", exist_ok=True)
os.makedirs("./deployment_repo/model_binary", exist_ok=True)

# =====================================================================
# 1. BUILD VECTOR DATABASE (synthetic embeddings -> exhaustive records)
# =====================================================================
print("[1/5] Restructuring numeric data into a searchable vector store...")

np.random.seed(42)
DIM    = 128
N_DOCS = 10_000

# Simulated corpus: short labelled text chunks paired with embeddings.
# In production these vectors would come from a real embedding model
# (e.g. sentence-transformers); here we generate them so the search
# layer below can be benchmarked deterministically.
corpus_labels = [f"doc_{i:05d}" for i in range(N_DOCS)]
database      = np.random.rand(N_DOCS, DIM).astype(np.float32)

vector_blueprint = {
    "embedding_dim"  : DIM,
    "total_vectors"  : N_DOCS,
    "distance_metric": "cosine",
    "sample_ids"     : corpus_labels[:5],
}
with open("./deployment_repo/model_metadata/vector_store_blueprint.json", "w") as f:
    json.dump(vector_blueprint, f, indent=2)

# =====================================================================
# 2. NORMALIZE + INDEX (the "optimization" layer)
# =====================================================================
print("[2/5] Normalizing vectors and building the fast-search index...")

def normalize(vectors: np.ndarray) -> np.ndarray:
    """L2-normalize so dot product becomes cosine similarity (no magnitude division at query time)."""
    norms = np.linalg.norm(vectors, axis=1, keepdims=True)
    norms[norms == 0] = 1
    return vectors / norms

norm_db = normalize(database)

# Optional: FAISS acceleration if available, otherwise fall back to
# the normalized-dot-product matrix multiply (still very fast on CPU).
try:
    import faiss
    index = faiss.IndexFlatIP(DIM)     # inner product on normalized vecs == cosine sim
    index.add(norm_db)
    backend = "faiss.IndexFlatIP"
except ImportError:
    index   = None
    backend = "numpy matmul (faiss not installed)"

print(f"  Search backend: {backend}")

# =====================================================================
# 3. CORE SEARCH FUNCTION
# =====================================================================
print("[3/5] Defining the optimized similarity-search routine...")

def vector_search(query_vec: np.ndarray, top_k: int = 5):
    """Return the top_k closest corpus entries to query_vec, with timing."""
    q = normalize(query_vec.reshape(1, -1))[0]
    start = time.time()

    if index is not None:
        scores, idxs = index.search(q.reshape(1, -1), top_k)
        results = list(zip(idxs[0].tolist(), scores[0].tolist()))
    else:
        scores  = norm_db @ q
        top_idx = np.argsort(scores)[::-1][:top_k]
        results = [(int(i), float(scores[i])) for i in top_idx]

    elapsed_ms = (time.time() - start) * 1000
    return results, elapsed_ms

# Quick benchmark vs naive brute-force loop, for the report below
def brute_force_search(query_vec: np.ndarray, top_k: int = 5):
    start = time.time()
    sims  = [(i, float(np.dot(query_vec, v) /
                       (np.linalg.norm(query_vec) * np.linalg.norm(v) + 1e-9)))
             for i, v in enumerate(database)]
    sims.sort(key=lambda x: x[1], reverse=True)
    elapsed_ms = (time.time() - start) * 1000
    return sims[:top_k], elapsed_ms

sample_query              = np.random.rand(DIM).astype(np.float32)
opt_results, opt_time_ms  = vector_search(sample_query, top_k=5)
bf_results,  bf_time_ms   = brute_force_search(sample_query, top_k=5)

print(f"  Brute-force search : {bf_time_ms:.2f} ms")
print(f"  Optimized search   : {opt_time_ms:.2f} ms  "
      f"({bf_time_ms / max(opt_time_ms, 1e-6):.1f}x faster)")

# =====================================================================
# 4. DEFINE KERNEL-LEVEL SEARCH BRIDGE (NO BACKEND CODE NEEDED)
# =====================================================================
print("[4/5] Binding native Colab execution hooks...")

def colab_search_bridge(user_msg):
    """Executes a similarity search directly inside the current notebook runtime memory space."""
    try:
        # Treat the chat input as a seed to deterministically derive a
        # query vector (stand-in for "embed this text" in a real pipeline).
        seed   = abs(hash(user_msg)) % (2**32)
        rng    = np.random.RandomState(seed)
        q_vec  = rng.rand(DIM).astype(np.float32)

        results, elapsed_ms = vector_search(q_vec, top_k=5)
        lines = [f"Query: '{user_msg}'  →  search time {elapsed_ms:.2f} ms ({backend})", ""]
        for rank, (idx, score) in enumerate(results, 1):
            lines.append(f"  {rank}. {corpus_labels[idx]}   similarity={score:.4f}")
        clean_reply = "\n".join(lines)
    except Exception as e:
        clean_reply = f"Search error: {e}"

    return IPython.display.JSON({"response": clean_reply})

output.register_callback('notebook.execute_llm', colab_search_bridge)

# =====================================================================
# 5. INJECT INDEPENDENT LOCAL FRONTEND WORKSPACE
# =====================================================================
print("[5/5] Displaying user interface...")

chat_ui_html = """
<div id="dsa-app-wrapper">
    <div class="chat-header">
        <h2>Vector Search Bench</h2>
        <p>Type anything — it's hashed into a query vector and searched against a 10,000-doc index.</p>
    </div>
    <div class="message-area" id="msgArea">
        <div class="msg bot-msg">System Online. Type any text to trigger a live similarity search.</div>
    </div>
    <div class="input-area">
        <input type="text" id="userInput" placeholder="Type a query..." onkeydown="if(event.key === 'Enter') sendQuery()">
        <button id="sendBtn" onclick="sendQuery()">Search</button>
    </div>
</div>

<style>
    #dsa-app-wrapper {
        background: #1e293b; color: #f8fafc; border-radius: 12px; max-width: 800px;
        margin: 20px auto; padding: 20px; display: flex; flex-direction: column; height: 550px;
        font-family: system-ui, sans-serif; border: 1px solid #334155; box-shadow: 0 4px 15px rgba(0,0,0,0.2);
    }
    .chat-header h2 { margin: 0; font-size: 1.4rem; color: #3b82f6; }
    .chat-header p { margin: 5px 0; font-size: 0.85rem; color: #94a3b8; }
    .message-area { flex: 1; padding: 15px; overflow-y: auto; display: flex; flex-direction: column; gap: 12px; background: #0f172a; border-radius: 8px; }
    .msg { max-width: 85%; padding: 12px; border-radius: 8px; font-size: 0.95rem; line-height: 1.6; white-space: pre-wrap; }
    .user-msg { background: #3b82f6; color: white; align-self: flex-end; border-bottom-right-radius: 0px; }
    .bot-msg { background: #334155; color: #f8fafc; align-self: flex-start; border: 1px solid #475569; border-bottom-left-radius: 0px; }
    .input-area { margin-top: 15px; display: flex; gap: 10px; }
    #userInput { flex: 1; background: #0f172a; border: 1px solid #475569; border-radius: 6px; padding: 12px; color: white; outline: none; font-size: 0.95rem; }
    #userInput:focus { border-color: #3b82f6; }
    #sendBtn { background: #3b82f6; color: white; border: none; padding: 0 24px; border-radius: 6px; cursor: pointer; font-weight: bold; font-size: 0.95rem; }
    #sendBtn:hover { background: #2563eb; }
</style>

<script>
    async function sendQuery() {
        const inputField = document.getElementById("userInput");
        const sendBtn = document.getElementById("sendBtn");
        const msgArea = document.getElementById("msgArea");
        const queryText = inputField.value.trim();
        if (!queryText) return;

        appendBubble(queryText, "user-msg");
        inputField.value = "";
        inputField.disabled = true; sendBtn.disabled = true;

        const loadingId = appendBubble("Embedding query and searching the index...", "bot-msg");

        try {
            google.colab.kernel.invokeFunction('notebook.execute_llm', [queryText], {}).then(function(result) {
                const responseData = result.data['application/json'];
                document.getElementById(loadingId).innerText = responseData.response;
            });
        } catch (error) {
            document.getElementById(loadingId).innerText = "Runtime connection error. Unable to reach search bridge.";
        } finally {
            inputField.disabled = false; sendBtn.disabled = false;
            inputField.focus();
            msgArea.scrollTop = msgArea.scrollHeight;
        }
    }

    function appendBubble(text, className) {
        const msgArea = document.getElementById("msgArea");
        const div = document.createElement("div");
        const id = "m_" + Date.now() + Math.floor(Math.random()*1000);
        div.id = id; div.className = "msg " + className; div.innerText = text;
        msgArea.appendChild(div);
        msgArea.scrollTop = msgArea.scrollHeight;
        return id;
    }
</script>
"""

IPython.display.display(IPython.display.HTML(chat_ui_html))
print("\nInterface active above. Use the search pane directly without network setup requirements.")
```
