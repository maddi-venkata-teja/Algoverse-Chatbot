```python
from langchain_ollama import ChatOllama
from langgraph.prebuilt import create_react_agent
from tools import ALL_TOOLS
```

```python
model = ChatOllama(
    model="llama3.1:8b",   # change to "qwen2.5:3b" if that's what you pulled
    temperature=0,
)
```

```python
agent = create_react_agent(model, ALL_TOOLS)
```

```python
def ask(query: str):
    result = agent.invoke({"messages": [{"role": "user", "content": query}]})
    final_message = result["messages"][-1]
    print(final_message.content)
    return result
```

```python
ask("What is (45 + 15) * 3?")
```

```python
ask("Save a note that says 'follow up on dataset filtering', then list all my notes.")
```

```python
ask("What's the weather like in Hyderabad?")
```

```python
ask("Generate a comment for this code: def add(a, b): return a + b")
```

```python
result = agent.invoke({"messages": [{"role": "user", "content": "What is 12 * 12?"}]})
for msg in result["messages"]:
    print(f"[{msg.type}] {msg.content}")
```
