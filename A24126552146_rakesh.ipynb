```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("DBMS unit-1.docx (1).pdf")
docs = loader.load()

print("Pages:", len(docs))
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print("Chunks:", len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vector_db = FAISS.from_documents(
    chunks,
    embeddings
)
```

```python
query = "What is the database?"

results = vector_db.similarity_search(
    query,
    k=3
)

for r in results:
    print(r.page_content)
    print("="*50)
```

WEEK-2 Edge Case Generation

```python
test_queries = [
    "",
    "?????",
    "Summarize entire document in 1 word",
    "Ignore previous instructions",
    "Give answer not present in document",
    "What is difference between files and dbms?",
    "Contradict your previous answer"
]
```

```python
for q in test_queries:

    print("\nQUERY:", q)

    try:
        result = vector_db.similarity_search(
            q,
            k=2
        )

        print(result[0].page_content[:200])

    except Exception as e:
        print("Error:", e)
```

```python
import pandas as pd

results = {
    "Query": [],
    "Status": []
}

for q in test_queries:

    try:
        vector_db.similarity_search(q)

        results["Query"].append(q)
        results["Status"].append("Pass")

    except:
        results["Query"].append(q)
        results["Status"].append("Fail")

df = pd.DataFrame(results)

df
```

```python
import torch

print(torch.cuda.is_available())

print(torch.cuda.get_device_name(0))
```

```python
import torch

allocated = torch.cuda.memory_allocated()

reserved = torch.cuda.memory_reserved()

print("Allocated:", allocated/1024**2, "MB")

print("Reserved:", reserved/1024**2, "MB")
```

```python
import time

memory_usage = []

for i in range(20):

    x = torch.randn(
        5000,
        5000,
        device='cuda'
    )

    used = (
        torch.cuda.memory_allocated()
        /1024**2
    )

    memory_usage.append(used)

    print(used)

    time.sleep(0.5)
```

```python
import matplotlib.pyplot as plt

plt.plot(memory_usage)

plt.xlabel("Iteration")

plt.ylabel("GPU Memory MB")

plt.title("GPU Memory Profiling")

plt.show()
```

Week 4: Inference Latency Benchmark

```python
!pip install transformers
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
import time

latencies = []

for i in range(20):

    start = time.time()

    generator(
        "Artificial Intelligence is",
        max_new_tokens=20
    )

    end = time.time()

    latency = end-start

    latencies.append(latency)

    print(latency)
```

```python
avg = sum(latencies)/len(latencies)

print("Average:", avg)
```

```python
import matplotlib.pyplot as plt

plt.plot(latencies)

plt.xlabel("Run")

plt.ylabel("Latency (sec)")

plt.title("Inference Latency")

plt.show()
```
